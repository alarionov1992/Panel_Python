# Sleep equation
Датасет `sleep75`

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import seaborn as sns
```

Импортируем данные

```python
sleep_df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/sleep75.csv')
sleep_df
```

```python
# Размер датасет
sleep_df.shape
```

Визуализация данных

```python
# Гистограмма
sns.histplot(data=sleep_df, x='sleep')
```

```python
sns.histplot(data=sleep_df, x='totwrk')
```

```python
sns.scatterplot(data=sleep_df, x='totwrk', y='sleep', hue='male')
```

Визуализация с подогнанной прямой

```python
sns.regplot(data=sleep_df, x='totwrk', y='sleep', ci=False, marker='.', line_kws={'color':'C1'})
```

Рассмотрим модель регрессии $sleep=\beta_0+\beta_1totwrk+\beta_2age+u$

```python
mod = smf.ols(formula='sleep~totwrk+age', data=sleep_df).fit()
mod.params.round(2)
```
