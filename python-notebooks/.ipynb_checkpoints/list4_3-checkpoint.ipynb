```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import scipy
```

##### Задача 1.1

```python
data1 = pd.read_csv('data-csv/sleep75.csv')
data1.shape
```

```python
# Создаем спецификацию модели через формулу
mod1 =smf.ols(formula='sleep~totwrk+age+south+male+smsa+yngkid+marr+union', data=data1)
#Подгоняем модель
Sleep_eq1 = mod1.fit()
```

```python
# отчет представляет собой набор из трёх таблиц(нумерация начинается с нулевой). 
# Нас интересуте вторая таблица(под индексом 1)
Sleep_eq1.summary()
```

```python
# Отобразим интересующую нас таблицу
table = Sleep_eq1.summary().tables[1]
table
```

```python
# Извлечь данные из данной таблицы можно вручную (Ctrl - C, Ctrl - V) и получить строку, чтобу преобразовать её в DataFrame
# А можно сохранить в виде html файла и уже его преобразовать в DataFrame
df = pd.read_html(Sleep_eq1.summary().tables[1].as_html(), header=0, index_col=0)[0] # Эту строку запоминать не нужно
df
```

```python
# Дополним DataFrame признаком t-values
df['tvalues']  = df['coef']/df['std err']
df
```

```python
# Сравним в полученными значениями с помощью регрессионной модели
# t-значение для коэффициентов при регрессорах с округлением
Sleep_eq1.tvalues.round(3)
```

1.2

```python
# Создаем спецификацию модели через формулу
mod2 = smf.ols(formula=' sleep~totwrk+age+south+male+smsa+yngkid+marr+union+np.log(hrwage)', data=data1)
#Подгоняем модель
Sleep_eq2 = mod2.fit()
```

```python
# отчет
Sleep_eq2.summary().tables[1]
```

### Проверка значимости коэффциентов регрессии Sleep_eq2

### Тестируем гипотезу $H_0:\beta_{j}=0$ VS $H_1:\beta_{j}\ne 0$ 

#### В отчете печатается следующая информация

#### в столбце $coef$: $\hat{\beta}_{j}$

#### в столбце $std \quad err$: $s.e.(\beta_{j})$

#### в столбце $t$:  $t$-value, $\displaystyle t=\frac{\hat{\beta_j}}{s.e.(\beta_j)}$

#### в столбце P>|t|: $p$-value

```python
alpha = 0.1 # уровень значимости
print('Уровень значимости alpha: ', alpha)
DF = int(Sleep_eq2.df_resid) # число степеней свободы 
t_crit = scipy.stats.t.ppf(1-alpha/2, DF)
print('Критическое значение t_crit: ', t_crit.round(3))
```

```python
df['znachimosty'] = df.apply(lambda x : 'znachim' if np.abs(x['tvalues']) > t_crit else 'ne znachim', axis = 1 )
df
```

##### Задача 2.1

```python
data2 = pd.read_csv('data-csv/Labour.csv')
data2.shape
```

```python
# Создаем спецификацию модели через формулу
mod3 = smf.ols(formula='np.log(output)~np.log(capital)+np.log(labour)', data=data2)
#Подгоняем модель
Output_eq1 = mod3.fit()
```

```python
Output_eq1.summary()
```

```python
alpha = 0.01
print('Уровень значимости alpha: ', alpha)
DF = int(Output_eq1.df_resid)
t_crit = scipy.stats.t.ppf(1-alpha/2, DF)
print('Критическое значение t_crit: ', t_crit.round(3))
```

### Тестируем гипотезу $H_0:\beta_{capital}=0.5$ VS $H_0:\beta_{capital}\ne 0.5$

### Петатается $\hat{\beta}_{capital}$, $s.e.(\beta_{capital})$, $\displaystyle t=\frac{\hat{\beta}-0.5}{s.e.(\beta)}$

```python
# формулируем гипотезу
hypotheses = 'np.log(capital)=0.5' # Запись означает: коэффициент при регрессоре log(capital) равен 0.5
# проверяем гипотезу
Output_eq1.t_test(hypotheses)
```

### ВЫВОД: Так как $|t-value| > t-crit$ => отвергаем $H_0$ => коэффициент при регрессоре $log(capital)$ значимо отличается от числа 0.5
