```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/ryupepa/Econometrica/master/python-notebooks/data-csv/sleep75.csv')
df
```

```python
model = smf.ols(data=df, formula='sleep~totwrk+age+male+smsa+south').fit()
# Параметры модели совпадают с Вашими
model.params
```

```python
# Дополним датасет предсказанными значениями
df['slp2'] = model.predict().round(2)**2
df['slp3'] = model.predict().round(2)**3
df['slp4'] =model.predict().round(2)**4
df
```

```python
# Строим вспомогательную регрессию, которая содержить степени предсказанных значений
models_unrest=smf.ols(data=df,formula='sleep~totwrk+age+male+smsa+south+slp2+slp3+slp4').fit()
models_unrest.params
```

```python
# Проверяем значимость коэффициентов при предсказанных значениях
models_unrest.f_test('slp2=slp3=slp4=0')
```

```python
R2_res=model.rsquared
R2_unres = models_unrest.rsquared
```

```python
((R2_unres - R2_res)/(1-R2_unres))*((len(df)-9)/3)
```
